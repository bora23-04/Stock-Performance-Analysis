### Data Cleaning

```python
tcs["Date"] = pd.to_datetime(tcs["Date"])
hcl["Date"] = pd.to_datetime(hcl["Date"])
infy["Date"] = pd.to_datetime(infy["Date"])
```

```python
cols = ["Adj Close", "Close", "High", "Low", "Open", "Volume"]

tcs[cols] = tcs[cols].apply(pd.to_numeric)
hcl[cols] = hcl[cols].apply(pd.to_numeric)
infy[cols] = infy[cols].apply(pd.to_numeric)
```

### Data Merging

```python
tcs["Company"] = "TCS"
hcl["Company"] = "HCL"
infy["Company"] = "INFY"
```

```python
final_df = pd.concat([tcs, hcl, infy])
final_df.reset_index(drop=True, inplace=True)
```

```python
final_df.head(10)
```

### Feature engineering

###### Measures profit/loss over time

```python
final_df["Daily Return"] = final_df.groupby("Company")["Close"].pct_change() * 100
```

###### MA_7(7 day Moving Average) needs 7 values and MA_30(30 days Moving Averages) needs 30 values

```python
final_df["MA_7"] = final_df.groupby("Company")["Close"] \
    .rolling(7).mean().reset_index(level=0, drop=True)

final_df["MA_30"] = final_df.groupby("Company")["Close"] \
    .rolling(30).mean().reset_index(level=0, drop=True)
```

###### Measures how much price fluctuates

```python
final_df["Volatility"] = final_df["High"] - final_df["Low"]
```

```python
final_df["Month"] = final_df["Date"].dt.month
final_df["Year"] = final_df["Date"].dt.year
```

###### Shows total growth over time

```python
final_df["Cumulative Return"] = final_df.groupby("Company")["Daily Return"] \
    .transform(lambda x: (1 + x/100).cumprod())
```

###### Shows actual gain/loss in Rs.

```python
final_df["Price Change"] = final_df.groupby("Company")["Close"].diff()
```

###### High-Low Percentage (Volatility %)

```python
final_df["HL_Pct"] = ((final_df["High"] - final_df["Low"]) / final_df["Close"]) * 100
```

###### Based on MA crossover

```python
final_df["Trend"] = final_df["MA_7"] > final_df["MA_30"]
```

###### Exponential Moving Average (EMA)

```python
final_df["EMA_7"] = final_df.groupby("Company")["Close"].transform(lambda x: x.ewm(span=7, adjust=False).mean())
```

```python
final_df.groupby("Company")["Trend"].mean()
```

###### HCL stays in uptrend more frequently

### INSIGHT:
